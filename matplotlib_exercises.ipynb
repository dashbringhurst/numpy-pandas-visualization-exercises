```python
# Use matplotlib to plot the following equation:
# y = x**2 − x + 2
# You'll need to write the code that generates the x and y points.

# Add an anotation for the point 0, 0, the origin.

x = range(-50, 50)
y = [n**2 - n + 2 for n in x]
plt.plot(x,y)
plt.annotate('origin', xy=(0,0), xytext=(-5,500), arrowprops={'facecolor':'purple'})
plt.show()
```

```python
# Combine the figures you created in the last step into one figure where each of the 4 equations has a different
# color for the points. Be sure to include a legend and an appropriate title for the figure.

plt.figure(figsize=(12,10))
plt.ylim(-5,20)
plt.xlim(-2,5)

x = range(0,10)
y = [n**.5 for n in x]
plt.plot(x,y, label='x^1/2', ls='--')

x = range(-10,10)
y = [n**3 for n in x]
plt.plot(x,y, label='x^3', ls=':')

x = range(0,10)
y = [2**n for n in x]
plt.plot(x,y, label='2^x')

x = range(0,10)
y = [1/(n+1) for n in x]
plt.plot(x,y, label='1/(x+1)')

plt.legend(loc='upper left')
plt.title('Four types of equations')
plt.xlabel('$x$')
plt.ylabel('$y$', rotation=0)

plt.show()
```
